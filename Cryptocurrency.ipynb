#### How the price of Solana has changed over time.
- Using plt to create a bar graph that shows the price of Solana over time.

```python
plt.figure(figsize=(13,6))
plt.bar(solana_set["Date"], solana_set["Close"], width = 4)
plt.title("Change in price of Solana")
plt.xlabel("Date")
plt.ylabel("Price")
plt.gca().xaxis.set_major_locator(mdates.WeekdayLocator(interval=13))
```

#### How the price of Bitcoin has changed over time.
- Using plt to create a bar graph that shows the price of Bitcoin over time.

```python
plt.figure(figsize=(13,6))
plt.bar(bitcoin_set["Date"], bitcoin_set["Close"], width = 4)
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Change in price of Bitcoin")
plt.gca().xaxis.set_major_locator(mdates.WeekdayLocator(interval=75))
```

#### Renaming columns.
- Renaming the "Close" columns in both data sets to their corresponding symbols for simplicity purposes.

```python
solana_set = solana_set[["Date", "Close"]].rename(columns={"Close": "sol"})
bitcoin_set = bitcoin_set[["Date", "Close"]].rename(columns={"Close": "btc"})
```

#### Comparing the price of Solana and Bitcoin over time.
- Creating a combined graph to show how the prices compare over time together.

```python
combined = pd.merge(solana_set, bitcoin_set)
fig, ax1 = plt.subplots(figsize=(13,6))

ax1.plot(combined["Date"], combined["sol"], label = "Solana", color = "Orange")
ax1.set_ylabel("Solana Price", color = "Orange")
ax1.set_xlabel("Date")

ax2 = ax1.twinx()
ax2.plot(combined["Date"], combined["btc"], label = "Bitcoin", color = "Blue")
ax2.set_ylabel("bitcoin Price", color = "Blue")

plt.gca().xaxis.set_major_locator(mdates.WeekdayLocator(interval=15))
plt.title("Comparing Solana and Bitcoin prices")
```

##### Showing the correlation between BTC and SOL.

```python
correlation_chart = combined["sol"].corr(combined["btc"])
print(f"Correlation: {correlation_chart:.3f}")
```
